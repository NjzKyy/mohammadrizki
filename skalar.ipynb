```python
x = 1
print("Nilai x:", x)
print("Tipe data x:", type(x))

y = 1
print("Nilai y:", y)
print("Tipe data y:", type(y))

hasil_xy = x + y
print("Hasil x + y:", hasil_xy)

x_float = 3.5
print("Nilai x_float:", x_float)

hasil_float = x_float + y
print("Hasil x_float + y:", hasil_float)

print("Tipe data x_float:", type(x_float))
```

```python
import tensorflow as tf

# buat variabel skalar x dan simpan suatu nilai skalar
x = tf.constant(1)
print("Nilai x:", x)
print("Tipe data x:", type(x))

# buat variabel skalar y dan simpan suatu nilai skalar
y = tf.constant(1)
print("Nilai y:", y)
print("Tipe data y:", type(y))

# jumlahkan x + y
hasil_xy = tf.add(x, y)
print("Hasil x + y:", hasil_xy)

# buat variabel x_float dan simpan nilai desimal
x_float = tf.constant(3.5)
print("Nilai x_float:", x_float)

# jumlahkan x_float dengan y
hasil_float = tf.add(x_float, tf.cast(y, tf.float32))
print("Hasil x_float + y:", hasil_float)

# tampilkan tipe data dari x_float
print("Tipe data x_float:", type(x_float))
```

```python
import torch

# buat variabel skalar x dan simpan suatu nilai skalar
x = torch.tensor(1)
print("Nilai x:", x)
print("Tipe data x:", type(x))

# buat variabel skalar y dan simpan suatu nilai skalar
y = torch.tensor(1)
print("Nilai y:", y)
print("Tipe data y:", type(y))

# jumlahkan x + y
hasil_xy = x + y
print("Hasil x + y:", hasil_xy)

# buat variabel x_float dan simpan nilai desimal
x_float = torch.tensor(3.5)
print("Nilai x_float:", x_float)

# jumlahkan x_float dengan y
hasil_float = x_float + y
print("Hasil x_float + y:", hasil_float)

# tampilkan tipe data dari x_float
print("Tipe data x_float:", type(x_float))
```
