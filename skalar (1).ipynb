```python
import numpy as np

x = 1
y = 1

M_np = np.array([[x, y, x],
                 [y, x, y],
                 [x, y, x]])

fro_np = np.linalg.norm(M_np, 'fro')

print("Matrix NumPy:\n", M_np)
print("Frobenius Norm:", fro_np)
```

```python
import torch

x = 1.
y = 1.

M_pt = torch.tensor([[x, y, x],
                     [y, x, y],
                     [x, y, x]])

fro_pt = torch.norm(M_pt, p='fro')

print("Matrix PyTorch:\n", M_pt)
print("Frobenius Norm:", fro_pt)
```

```python
import tensorflow as tf

x = 1.
y = 1.

M_tf = tf.constant([[x, y, x],
                    [y, x, y],
                    [x, y, x]])

fro_tf = tf.norm(M_tf)

print("Matrix TensorFlow:\n", M_tf)
print("Frobenius Norm:", fro_tf.numpy())
```
